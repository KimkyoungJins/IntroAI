```python
import matplotlib.pyplot as plt


## 가로 10인치 세로 5인치 크기 지정하기
plt.figure(figsize =(10, 5))

## 1행 2열 중 첫번째(왼쪽)에 그릴 준비를 하여라
plt.subplot(1, 2, 1)
plt.xlim([-1., 1])    #x축의 범위를 -1 부터 1까지로 고정해 준다.
plt.ylim([-0.5, 2.5]) # y축의 범위를 ([-0.5. 2.5])


# 두번째 캔버스 
plt.subplot(1, 2, 2)
plt.ylabel('Y')
plt.xlabel('X')
plt.title("Hanse")
plt.show() # display the figure

```

```python
np.random.seed(500)

x = np.random.rand(50)
y = np.random.rand(50)


## creating scatter plotd을 만든다면
## x, y 좌표에 점을 찍는다. 
# 모든 점의 색을 검정으로 한다.
# alpha 0.5 투병도를 50퍼센트로 한다. 
plt.scatter(x, y, c = "black", alpha = 0.5)
plt.show()

```

```python
import matplotlib.pyplot as plt

year2018 = [1, 5, 8, 9, 7, 11, 8, 12, 14, 9, 5]
year2019 = [3, 7, 9, 6, 4, 5, 14, 7, 6, 16, 12]

##potting line plots
plt.plot(range(1, 12), year2018, color = 'black')
plt.plot(range(1, 12), year2019, color = 'red', alpha = 0.6)

## creating points
plt.plot(range(1, 12), year2018, 'o', color = 'orange', alpha = 0.5)
plt.plot(range(1, 12), year2019, 'o', color = 'blue', alpha = 0.4)

## 타이틀 데이터 정보 x, y축에 대하여서 설명하기
plt.xlabel('Months')
plt.ylabel('Values')
plt.title('Good day')
plt.legend(['year2019', 'year2018'], loc = 0)

plt.show()
```
