```python
import string
print('Punctuation: ', string.punctuation)
# Punctuation:  !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~


def remove_punctuation(text):
    sent = []
    for t in text.split(' '):
        # 각 단어(t)에 대해, 문자 하나하나(c)를 순회하면서
        # c가 punctuation 목록에 없으면 그대로 남기고,
        # 있으면 건너뛰어(join 시 빠짐)
        no_punct = "".join([c for c in t if c not in string.punctuation])
        sent.append(no_punct)

    # 구두점이 제거된 단어들을 다시 공백으로 연결해서 문장으로 복원
    sentence = " ".join(s for s in sent)
    return sentence
```
